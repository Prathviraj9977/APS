```python
# Exercise 1
import time

def func1(a, b):
    result = 0
    for i in range(0, b):
        result += a
    return result

start = time.time()          # record start time
output = func1(4, 3)         # run the function
end = time.time()            # record end time

print("Result:", output)
print("Execution time:", end - start, "seconds")
```

```python
# Exercise 2
import time
def func2(a, b):
  result = 1
  while b > 0:
    result = result * a
    b = b - 1
  return result

  start=time.time()
  output=func2(2,4)
  end=time.time()
  print("Result:", output)
print("Execution time:", end - start, "seconds")
```

```python
# Exercise 3
import time
def func3(num):
   result = 0
   while num > 0:
    result = result + num % 10
    num //= 10
    return result

start=time.time()
output=func3(1234)
end=time.time()
print("Result:", output)
print("Execution time:", end - start, "seconds")
```

```python
# Exercise 4
def func4(a, b):
    b.sort() # Time complexity for this function is O(Nlog N) where N = len(b)
    intersect = 0
    for x in a:
        if binary_search(b, x) >= 0:
            intersect = intersect + 1
    return intersect

def binary_search(arr, target):
    low, high = 0, len(arr) - 1
    while low <= high:
        mid = (low + high) // 2
        if arr[mid] == target:
            return mid
        elif arr[mid] < target:
            low = mid + 1
        else:
            high = mid - 1
    return -1


start=time.time()
output=func4([1, 2, 3, 4], [2, 4, 6, 8])
end=time.time()
print("Result:", output)
print("Execution time:", end - start, "seconds")


```

```python
# Exercise 6
def func8(A):
    func8_helper(A, 0, len(A) -1)

def func8_helper(A, i, j):
    #Input: An array A and two indices i and j.
    n = j - 1 + 1
    if n ==2:
        if A[i] > A[j]:
            # Swap A[i] and A[j]
            A[i], A[j] = A[j], A[i]
        elif n > 2:
            m = n // 3
            func8_helper(A, i, j - m)
            func8_helper(A, i + m, j)
            func8_helper(A, i, j - m)
        return A

start=time.time()
output=func8([3,1,2,7,8,6])
end=time.time()
print("Result:", output)
print("Execution time:", end - start, "seconds")
```

```python
# Exercise 7
def func12(arr, k):
    arr_len = len(arr)
    k = k % arr_len
    # Initialize a new array of the same length as 'arr'
    custom_arr = [0] * arr_len

    for i in range(arr_len):
        mystery_index = (i + k) % arr_len
        custom_arr[mystery_index] = arr[i]
    return custom_arr

start=time.time()
output=func12([1, 2, 3, 4, 5], 2)
end=time.time()
print("Result:", output)
print("Execution time:", end - start, "seconds")

```

```python
#write a function named as max_min to find minimum (or maximum) element in an array
def max_min(arr):
  max=arr[0]
  min=arr[0]
  for i in range (len(arr)):
    if arr[i]>max:
      max=arr[i]
    elif arr[i]<min:
      min=arr[i]
  return min,max


arr=[45,5,3,2,98,1,23]
print("minimum and maximaum in array:",max_min(arr))
```

```python
# write a function named as Counting_EvenOdd to count number of even and odd elements in an array
def Counting_EvenOdd(arr):
  even =0
  odd=0
  for i in range(len(arr)):
    if arr[i]%2==0:
      even+=1
    else:
      odd+=1
  return even, odd

arr=[2,3,7,4,8,22,29,40]
print("Count Number of Even and odd Element in array:",Counting_EvenOdd(arr))
```
